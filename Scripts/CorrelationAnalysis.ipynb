```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\madara.premawardhana\Documents\UniversityCollegeDublin\QuickResearch\Datasets\US_Chicago_Lake_Michigan_lakefront_cleaned.csv')
print (df)
```

```python
fig, ax=plt.subplots(figsize=(10,6))
corr = df.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask,center=0, cmap='BrBG')

ax.set_title('Correlation of weather variables Chicago,USA', fontsize=18)
```
